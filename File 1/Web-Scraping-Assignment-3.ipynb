Q2. In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then scrape all the products available under that product name. Details to be scraped are: "Brand
Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and
“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
product_url = []
url = driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')

    
for i in url:
    product_url.append(i.get_attribute('href'))

price= []
rep= [] 
brand = []
delivery = []
availablity = []
product_name = []

for i in product_url:
    driver.get(i)
    time.sleep(5)
    try:
        prices = driver.find_element(By.XPATH,"/html/body/div[2]/div/div[5]/div[3]/div[4]/div[13]/div[4]/div[1]/span[2]/span[2]/span[2]")
        price.append(prices.text.replace('\n00',""))
    except NoSuchElementException:
        price.append('-')

    try:
        replacement = driver.find_element(By.XPATH,"/html/body/div[2]/div/div[5]/div[3]/div[4]/div[24]/div[2]/div/div/div/div[2]/div/ol/li[3]/div/span/div[2]/span")
        rep.append(replacement.text)
    except NoSuchElementException:
        rep.append('-')

    try:
        brands = driver.find_element(By.XPATH,"/html/body/div[2]/div/div[5]/div[3]/div[4]/div[47]/div/table/tbody/tr[1]/td[2]/span")
        brand.append(brands.text)
    except NoSuchElementException:
        brand.append('-')
        
    try:
        deli = driver.find_element(By.XPATH,"/html/body/div[2]/div[2]/div[5]/div[4]/div[1]/div[3]/div/div/div/div/form/div/div/div/div/div[3]/div/div[2]/div[8]/div[1]/div/div/div[1]/b")
        delivery.append(deli.text)
    except NoSuchElementException:
        delivery.append('Not Present')
        
    try:
        avail = driver.find_element(By.XPATH,"/html/body/div[2]/div/div[5]/div[3]/div[1]/div[3]/div/div[1]/div/div/div/form/div/div/div/div/div[4]/div/div[5]/div/div[1]/span")
        availablity.append(avail.text)
    except NoSuchElementException:
        availablity.append('-')
        
    try:
        name=driver.find_element(By.XPATH,"/html/body/div[2]/div/div[5]/div[3]/div[4]/div[1]/div/h1/span")
        prod_name=name.text
        product_name.append(prod_name)
        if(prod_name == 0):
            try:
                name1=driver.find_element(By.XPATH,'/html/body/div[2]/div[2]/div[6]/div[4]/div[4]/div[1]/div/h1/span')
                product_name.append(name1.text)
            except NoSuchElementException:
                product_name.append("-")
    except:
        product_name.append("-")
```

```python
print(len(brand),len(product_name),len(price),len(rep),len(delivery),len(availablity),len(product_url))
```

Q8. Write a program to extract at least 500 Comments, Comment upvote and time when comment was posted from any YouTube Video.

```python
driver = webdriver.Chrome()
time.sleep(2)

driver.get("https://www.youtube.com/watch?v=r-hZb0iVQZI")
time.sleep(3)
```

```python
youtube_video=pd.DataFrame({'Comments':comments,'Comments Upvote':upvote,'Comment is POSTED':posted})
youtube_video
```

Q9. Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in “London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall reviews, privates from price, dorms from price, facilities and property description.

```python
driver = webdriver.Chrome()
time.sleep(2)

driver.get("https://www.hostelworld.com/")
time.sleep(3)
```

```python
search_bar = driver.find_element(By.XPATH,'//input[@class="native-input"]')
search_bar.send_keys('London')
```

```python
button = driver.find_element(By.XPATH,'//button[@class="item-content"]')
button.click()
```

```python
search_button = driver.find_element(By.XPATH,'//button[@class="btn-content large-button"]')
search_button.click()
```

```python
hostel_name = []
distance = []
pvt_prices = []
dorms_price = []
rating = []
reviews = []
over_all = []
facilities = []
description =[]

# fetching the distance from city centre    
dist = driver.find_elements(By.XPATH,'//span[@class="distance-description"]')
try:
    for i in dist:
        distance.append(i.text.replace('Hostel - ',''))
except NoSuchElementException:
    distance.append('-')

    
product_url = []

p_url = driver.find_elements(By.XPATH,'//a[@class="property-card-container horizontal"]')
for i in p_url:
    product_url.append(i.get_attribute('href'))

for i in product_url[0:30]:
    driver.get(i)
    time.sleep(3)
    
# fetching hostel name
    try:
        name = driver.find_elements(By.XPATH,'//div[@class="title-2"]')
        for i in name:
            hostel_name.append(i.text)
    except NoSuchElementException:
        hostel_name.append('-')
        
    # fetching the property description 
    try:
        disc = driver.find_element(By.XPATH,'//div[@class="content collapse-content"]')
        description.append(disc.text)
    except NoSuchElementException:
        over_all.append('-')
        
    # fetching the ratings for the hostels
    try:
        rat = driver.find_element(By.XPATH,'//div[@class="score orange big"]')
        rating.append(rat.text)
    except NoSuchElementException:
        rating.append('-')
        
    # fetching the total reviews        
    try:
        rws = driver.find_element(By.XPATH,'//div[@class="reviews"]')
        reviews.append(rws.text.replace('Total Reviews',''))
    except NoSuchElementException:
        reviews.append('-')
        
    # fetching the overall review
    try:
        overall_rw = driver.find_element(By.XPATH,'//div[@class="keyword"]')
        over_all.append(overall_rw.text)
    except NoSuchElementException:
        over_all.append('-')
        
    # fetching the privates from price details
    try:
        pvt_price = driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/section/div[12]/div/div/section/div/div[1]/div[2]/div[1]/div[2]/div/div[2]/div[2]/div/div[1]/div[1]/div')
        pvt_prices.append(pvt_price.text)
    except NoSuchElementException:
        pvt_prices.append('-')
        
    # fetching the dorms from price information
    try:
        dorms = driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/section/div[11]/div/div/section/div/div[1]/div/div/div[2]/div[1]/div[2]/div[2]/div/div[1]/div[2]/div[1]')
        dorms_price.append(dorms.text)
    except NoSuchElementException:
        dorms_price.append('-')
        
    # fetching all the facilities
    try:
        fac = driver.find_elements(By.XPATH,'//ul[@class="facilities"]')
        for i in fac:
            facilities.append(i.text)
    except NoSuchElementException:
        facilities.append('-')    
```

```python
data = list(zip(hostel_name,distance,rating,reviews,over_all,pvt_prices,dorms_price,facilities,description))       
df = pd.DataFrame(data, columns = ["Hostel name","Distance from city centre","ratings","Total reviews",
                                   "Overall review","Privates from price","Dorms from price","Facilities","Property Description"])
df
```

```python
lap_name = []
op_stm = []
display = []
processor = []
memory = []
weight = []
dimensions = []
graph_processor = []
prices = []


name = driver.find_elements(By.XPATH,'//h3[@class="font130 mt0 mb10 mobilesblockdisplay "]')
for i in name:
    lap_name.append(i.text)
    

try:
    os = driver.find_elements(By.XPATH,'//*[@id="section-additional_information"]/div[2]/table/tbody/tr[2]/td/table/tbody/tr[1]/td/p')
    for i in os:
        op_stm.append(i.text)
except NoSuchElementException:
    pass

try:
    disp = driver.find_elements(By.XPATH,'//*[@id="section-additional_information"]/div[2]/table/tbody/tr[2]/td/table/tbody/tr[1]/td/p')
    for i in disp:
        display.append(i.text)
except NoSuchElementException:
    pass

try:
    pro = driver.find_elements(By.XPATH,'//*[@id="section-additional_information"]/div[2]/table/tbody/tr[12]/td/table/tbody/tr[2]/td/p')
    for i in pro:
        processor.append(i.text)
except NoSuchElementException:
    pass

try:
    mem = driver.find_elements(By.XPATH,'//*[@id="section-additional_information"]/div[2]/table/tbody/tr[14]/td/table/tbody/tr[1]/td/p')
    for i in mem:
        memory.append(i.text)
except NoSuchElementException:
    pass

try:
    wgt = driver.find_elements(By.XPATH,'//*[@id="section-additional_information"]/div[2]/table/tbody/tr[10]/td/table/tbody/tr[2]/td/p')
    for i in wgt:
        weight.append(i.text)
except NoSuchElementException:
    pass

try:
    dim = driver.find_elements(By.XPATH,'//*[@id="section-additional_information"]/div[2]/table/tbody/tr[10]/td/table/tbody/tr[1]/td/p')
    for i in dim:
        dimensions.append(i.text)
except NoSuchElementException:
    pass

try:
    g_pro = driver.find_elements(By.XPATH,'//*[@id="section-additional_information"]/div[2]/table/tbody/tr[12]/td/table/tbody/tr[3]/td/p')
    for i in g_pro:
        graph_processor.append(i.text)
except NoSuchElementException:
    pass

# fetching information for price
try:
    price = driver.find_elements(By.XPATH,'//div[@class="cegg-price cegg-price-color cegg-price-instock"]')
    for i in price:
        prices.append(i.text.replace('₹','Rs '))
except NoSuchElementException:
    pass
```

```python
data = list(zip(lap_name,op_stm,display,processor,memory,weight,dimensions,graph_processor,prices))
df = pd.DataFrame(data, columns = ["Laptop Name","Operating System","Display","Processor","Memory",
                                   "Weight(Kg)","Dimensions","Graphic-Processor","Price"])
df
```
